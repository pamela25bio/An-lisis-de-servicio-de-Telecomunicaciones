## Análisis estadístico

### Formulación de hipótesis

**Hipótesis 1** Llamadas perdidas

**H₀ (nula):** El promedio de llamadas entrantes perdidas es igual entre operadores eficaces e ineficaces.

**H₁ (alternativa):** El promedio de llamadas entrantes perdidas es diferente entre operadores eficaces e ineficaces.

**Hipótesis 2** Llamadas salientes

**H₀:** El promedio de llamadas salientes es igual entre operadores eficaces e ineficaces.

**H₁:** El promedio de llamadas salientes es diferente entre operadores eficaces e ineficaces.

### Identificación de operadores ineficaces

```python
# Análisis de métricas de los operadores
operator_data = tel_data.groupby('operator_id').agg(
    total_calls=('calls_count', 'sum'),
    missed_calls=('is_missed_call', lambda x: (x == True).sum()),
    out_calls=('direction', lambda x: (x == 'out').sum()),
    in_calls=('direction', lambda x: (x == 'in').sum()),
    total_duration=('total_call_duration', 'sum'),
    duration_mean=('call_duration', 'mean')
).reset_index()

# Columna adicinal de porcentaje de llamadas perdidas
operator_data['lost_perc'] = (
    operator_data['missed_calls'] / operator_data['total_calls'] * 100
)

operator_data.head(10)
```

ANALISIS: Las primeras filas nos muestran algunos operadores con un porcentaje de llamadas perdidas de importancia, al igual que las llamadas salientes tienden a ser bajas.

```python
# Separación de operadores ineficaces

mean_duration_call = operator_data['duration_mean'].mean()
q25_out_calls = operator_data['out_calls'].quantile(0.25)

# Definición de métricas para operadores ineficaces
operator_data['inefficient_op'] = (
    (operator_data['lost_perc'] > 10) |
    (operator_data['duration_mean'] > 2 * mean_duration_call) |
    (operator_data['duration_mean'] < 0.5 * mean_duration_call) |
    (operator_data['out_calls'] <= q25_out_calls)
)

# Conteo de operadores eficaces e ineficaces
print(operator_data['inefficient_op'].value_counts())

operator_data[operator_data['inefficient_op']].head(10)
```

ANALISIS: Hay una diferencia significativa entre los grupos, siendo mucho mayor el número de operadores ineficaces.

```python
# Conteo de operadores por categoría
sns.countplot(data=operator_data, x='inefficient_op')
plt.title('Número de operadores eficaces vs ineficaces')
plt.show()

# Boxplot para lost_perc
sns.boxplot(data=operator_data, x='inefficient_op', y='lost_perc')
plt.title('Comparación % llamadas perdidas por categoría')
plt.show()

# Boxplot para out_calls
sns.boxplot(data=operator_data, x='inefficient_op', y='out_calls')
plt.title('Comparación llamadas salientes por categoría')
plt.show()

# Boxplot para duration_mean
sns.boxplot(data=operator_data, x='inefficient_op', y='duration_mean')
plt.title('Comparación duración promedio de llamada')
plt.show()
```

ANALISIS: En este punto ya se clasificaron los operadores, mostrando que el grupo de operadores ineficaces es mayor que los operadores eficaces. Las métricas elegidas para clasificarlos nos muestran claramente que hay un gran número de llamadas perdidas y el tiempo de duración de las llamadas es más corto en operadores ineficaces, aunque en duración de llamadas hay valores atípicos, con una duración mucho mayor que el promedio, mientras que los operadores eficaces se mantienen con el número de llamadas salientes mayor y baja pérdida de llamadas.

```python
# Pruebas estadísticas

# Definir los dos grupos
efficient = operator_data[operator_data['inefficient_op'] == False]
inefficient = operator_data[operator_data['inefficient_op'] == True]

# Función para testear normalidad, t-test o Mann-Whitney
def compare_metrics(metric):
    print(f"\nAnálisis para {metric}:")

    # Test normalidad
    stat_efficient, p_efficient = shapiro(efficient[metric])
    stat_inefficient, p_inefficient = shapiro(inefficient[metric])
    print(f"Normalidad eficaz p-value: {p_efficient:.10f}")
    print(f"Normalidad ineficaz p-value: {p_inefficient:.10f}")

    # Elegir prueba según normalidad
    if p_efficient > 0.05 and p_inefficient > 0.05:
        # Test t de Student
        stat, p = ttest_ind(efficient[metric], inefficient[metric], equal_var=False)
        print(f"Prueba t de Student p-value: {p:.4f}")
    else:
        # Test Mann-Whitney
        stat, p = mannwhitneyu(efficient[metric], inefficient[metric], alternative='two-sided')
        print(f"Prueba Mann-Whitney p-value: {p:.4f}")

    # Interpretación simple
    if p < 0.05:
        print(f"Resultado: Diferencia estadísticamente significativa en {metric}.")
    else:
        print(f"Resultado: No hay diferencia estadísticamente significativa en {metric}.")

# Ejecutar para llamadas perdidas y llamadas salientes
compare_metrics('lost_perc')
compare_metrics('out_calls')
```

ANALISIS: Se puede ver claramente que hay diferencias significativas en los grupos, lo que nos sugiere que hay una gran diferencia entre las llamadas perdidas y las llamadas salientes entre los operadores eficaces y los operadores ineficaces, probando que las hipótesis alternativas sugeridas son correctas, refutando las hipótesis nulas.
